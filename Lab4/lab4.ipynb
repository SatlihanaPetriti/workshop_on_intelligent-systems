## Импорты библиотек

```python
from sklearn.model_selection import train_test_split  # Для разделения данных на обучающую и тестовую выборки
from sklearn.preprocessing import StandardScaler, LabelEncoder  # Для нормализации и кодирования данных
from sklearn.svm import SVC  # Модель опорных векторов
from sklearn.neighbors import KNeighborsClassifier  # Модель к ближайших соседей
from sklearn.linear_model import LogisticRegression  # Логистическая регрессия
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix  # Метрики для оценки моделей
import pandas as pd  # Для работы с данными
```

## Загрузка данных

```python
# Загрузить датасет и указать имена столбцов
column_names = ['age', 'operation_year', 'positive_auxillary_nodes', 'survival_status']
data = pd.read_csv(file_path, header=None, names=column_names)
```

```python
column_names = ['age', 'operation_year', 'positive_auxillary_nodes', 'survival_status']
data = pd.read_csv(file_path, header=None, names=column_names)
# Разделение данных на признаки и целевую переменную
X = data[['age', 'operation_year', 'positive_auxillary_nodes']]
y = data['survival_status']
# Разделить данные на обучающую (70%) и тестовую (30%) выборки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
# Стандартизировать признаки (привести их к одному масштабу)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## Обучение и оценка модели опорных векторов

```python
svm_model = SVC()  # Создать модель
svm_model.fit(X_train, y_train)  # Обучить модель на обучающей выборке
y_pred_svm = svm_model.predict(X_test)  # Сделать предсказания на тестовой выборке
print("Классификатор опорных векторов")
print(f"Точность: {accuracy_score(y_test, y_pred_svm):.2f}")  # Печать точности модели
print("Отчет по классификации:")
print(classification_report(y_test, y_pred_svm))  # Печать отчета по метрикам классификации (precision, recall, F1-score)
print("Матрица ошибок:")
print(confusion_matrix(y_test, y_pred_svm))  # Печать матрицы ошибок
print("\n")
```

## Обучение и оценка модели k-NN

```python
knn_model = KNeighborsClassifier(n_neighbors=5)  # Создать модель k-NN с 5 соседями
knn_model.fit(X_train, y_train)  # Обучить модель на обучающей выборке
y_pred_knn = knn_model.predict(X_test)  # Сделать предсказания на тестовой выборке
print("Классификатор k-NN")
print(f"Точность: {accuracy_score(y_test, y_pred_knn):.2f}")  # Печать точности модели
print("Отчет по классификации:")
print(classification_report(y_test, y_pred_knn))  # Печать отчета по метрикам классификации
print("Матрица ошибок:")
print(confusion_matrix(y_test, y_pred_knn))  # Печать матрицы ошибок
print("\n")
```

## Обучение и оценка модели логистической регрессии

```python
logreg_model = LogisticRegression(max_iter=1000)  # Создать модель логистической регрессии с увеличенным количеством итераций для сходимости
logreg_model.fit(X_train, y_train)  # Обучить модель на обучающей выборке
y_pred_logreg = logreg_model.predict(X_test)  # Сделать предсказания на тестовой выборке
print("Логистическая регрессия")
print(f"Точность: {accuracy_score(y_test, y_pred_logreg):.2f}")  # Печать точности модели
print("Отчет по классификации:")
print(classification_report(y_test, y_pred_logreg))  # Печать отчета по метрикам классификации
print("Матрица ошибок:")
print(confusion_matrix(y_test, y_pred_logreg))  # Печать матрицы ошибок
print("\n")
```

```python
# Cловарь для хранения результатов
results = {
    "SVM": {
        "Точность": accuracy_score(y_test, y_pred_svm),
    },
    "k-NN": {
        "Точность": accuracy_score(y_test, y_pred_knn),
    },
    "Логистическая регрессия": {
        "Точность": accuracy_score(y_test, y_pred_logreg),
    }
}

# Определение лучшей модели по точности
best_model = max(results.items(), key=lambda item: item[1]["Точность"])
print(f"Лучшая модель: {best_model[0]} с точностью {best_model[1]['Точность']:.2f}")
```
