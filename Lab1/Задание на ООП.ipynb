## Задание на ООП
- Класс Дробное число со знаком (Fractions).

Число должно быть представлено двумя полями: целая часть - длинное целое со знаком,
дробная часть - беззнаковое короткое целое. Реализовать арифметические операции сложения,  вычитания, умножения и операции сравнения.

В функции main проверить эти методы.

```python
class Fraction:
  # Инициализация объекта Fraction- param intg_part: Целая часть дробного числа (целое число),
  # param frac_part: Дробная часть дробного числа
    def __init__(self, intg_part: int, frac_part: int):
        self.intg_part = intg_part # Сохранение целой части дроби
        self.frac_part = frac_part # Сохранение дробной части дроби

    def to_float(self):
       # Определение знака целой части (если отрицательное, то sign = -1, иначе sign = 1)
        sign = -1 if self.intg_part < 0 else 1
       # Возвращение значения дроби в формате float, учитывая знак целой части и дробную часть
        return self.intg_part + sign * (self.frac_part / 1000)

    def __add__(self, other):
        """Сложение."""
        result = self.to_float() + other.to_float()  # Преобразование обеих дробей в формат float и сложение их
         # Извлечение целой части и вычисление дробной части
        intg_part = int(result)
        frac_part = round(abs(result - intg_part) * 1000)
        return Fraction(intg_part, frac_part)  # Возвращение нового объекта Fraction, представляющего результат сложения

    def __sub__(self, other):
        """Вычитание."""
        result = self.to_float() - other.to_float()
        intg_part = int(result)
        frac_part = round(abs(result - intg_part) * 1000)
        return Fraction(intg_part, frac_part)

    def __mul__(self, other):
        """Умножение двух дробей."""
        result = self.to_float() * other.to_float()
        intg_part = int(result)
        frac_part = round(abs(result - intg_part) * 1000)
        return Fraction(intg_part, frac_part)

    def __eq__(self, other):
        """Сравнение: равно."""
        return self.to_float() == other.to_float()

    def __lt__(self, other):
        """Сравнение: меньше."""
        return self.to_float() < other.to_float()

    def __le__(self, other):
        """Сравнение: меньше или равно."""
        return self.to_float() <= other.to_float()

    def __gt__(self, other):
        """Сравнение: больше."""
        return self.to_float() > other.to_float()

    def __ge__(self, other):
        """Сравнение: больше или равно."""
        return self.to_float() >= other.to_float()

    def __str__(self):
        """Возвращает строковое представление дробного числа."""
        return f"{self.intg_part}.{self.frac_part:03d}"
```

```python


def main():
    # Создание дробных чисел
    frac1 = Fraction(5, 240)
    frac2 = Fraction(-2, 700)
    print("f1:", frac1)
    print("f2:", frac2)
     # Арифметические операции с дробными числами
    print("Сложение:", frac1 + frac2)
    print("Вычитание:", frac1 - frac2)
    print("Умножение:", frac1 * frac2)

     # Операции сравнения между дробными числами
    print("Равно (==):", frac1 == frac2) # Проверка на равенство дробей
    print("Меньше (<):", frac1 < frac2)  # Проверка, является ли первая дробь меньше второй
    print("Меньше или равно (<=):", frac1 <= frac2) # Проверка, является ли первая дробь меньше или равна второй
    print("Больше (>):", frac1 > frac2) # Проверка, является ли первая дробь больше второй
    print("Больше или равно (>=):", frac1 >= frac2)  # Проверка, является ли первая дробь больше или равна второй



if __name__ == "__main__":
    main()
```
