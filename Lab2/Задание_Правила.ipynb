## Правила

Наше приложение будет решать вымышленную задачу, которую можно сформулировать при помощи следующих условий.
Компания под названием XYZ производит два типа компьютеров: «тип 1» и «тип 2». Тип компьютера определяется его архитектурой.
Компьютеры фирмы XYZ могут выполнять ряд функций. На данный момент определены четыре функции: сервер DDNS (DDNS Server), сервер DNS (DNS Server), шлюз (Gateway) и маршрутизатор (Router).
Каждый компьютер перед выпуском проходит ряд тестов.
Тесты, выполняемые над каждым компьютером, зависят от его типа и выполняемой функции. На данный момент определены пять тестов: «тест 1», «тест 2», «тест 3», «тест 4» и «тест 5».
Для каждого проверяемого компьютера устанавливается предельный срок тестирования. Все тесты, соответствующие данному компьютеру, должны быть выполнены не позднее указанной даты. Сама дата зависит от тестов, выбранных для каждого конкретного компьютера.
Большая часть процесса выполнения тестов в компании XYZ автоматизирована при помощи внутреннего программного обеспечения, которое выбирает конкретный набор тестов и определяет дату тестирования на основе типа и функций компьютеров.

Правила

На данный момент наборы тестов и даты их выполнения для конкретных типов компьютеров выбираются в соответствии со следующими бизнес-правилами.
Над компьютерами типа 1 должны быть выполнены тесты 1, 2 и 3.
Над компьютерами типа 2, выполняющими функцию серверов DNS, должны быть выполнены тесты 4 и 5.
Над компьютерами типа 2, выполняющими функцию серверов DDNS, должны быть выполнены тесты 2 и 3.
Над компьютерами типа 2, выполняющими функцию шлюза, должны быть выполнены тесты 3 и 4.
Над компьютерами типа 2, выполняющиими функцию маршрутизатора, должны быть выполнены тесты 1 и 3.
Если среди тестов, выбранных для данного компьютера, есть тест 1, то тестирование должно производиться не позднее чем через три дня после даты производства. Данное правило является приоритетным по отношению ко всем последующим правилам выбора даты тестирования.
Если среди тестов, выбранных для данного компьютера, есть тест 2, то тестирование должно производиться не позднее чем через семь дней после даты производства. Данное правило является приоритетным по отношению ко всем последующим правилам выбора даты тестирования.
Если среди тестов, выбранных для данного компьютера, есть тест 3, то тестирование должно производиться не позднее чем через 10 дней после даты производства. Данное правило является приоритетным по отношению ко всем последующим правилам выбора даты тестирования.
Если среди тестов, выбранных для данного компьютера, есть тест 4, то тестирование должно производиться не позднее чем через 12 дней после даты производства. Данное правило является приоритетным по отношению ко всем последующим правилам выбора даты тестирования.
Если среди тестов, выбранных для данного компьютера, есть тест 5, то тестирование должно производиться не позднее чем через 14 дней после даты производства.

**Архитектура компьютера:**
- Тип 1
- Тип 2

**Роли:**
- Сервер DDNS (DDNS)
- Сервер DNS (DNS)
- Шлюз (Gateway)
- Маршрутизатор (Router)

**Тесты:**
- Тест 1
- Тест 2
- Тест 3
- Тест 4
- Тест 5


**Правила тестирования:**
- Для компьютеров типа 1:
    - Должны быть выполнены Тест 1, Тест 2 и Тест 3.
- Для компьютеров типа 2:
    - Сервер DDNS: Должны быть выполнены Тест 2 и Тест 3.
    - Сервер DNS: Должны быть выполнены Тест 4 и Тест 5.
    - Шлюз: Должны быть выполнены Тест 3 и Тест 4.
    - Маршрутизатор: Должны быть выполнены Тест 1 и Тест 3.

**Правила для дедлайнов тестов:**

- Тест 1: Должен быть завершён не позднее 3 дней с момента производства.
- Тест 2: Должен быть завершён не позднее 7 дней с момента производства.
- Тест 3: Должен быть завершён не позднее 10 дней с момента производства.
- Тест 4: Должен быть завершён не позднее 12 дней с момента производства.
- Тест 5: Должен быть завершён не позднее 14 дней с момента производства.
- Эти дедлайны приоритетны по порядку: Тест 1 → Тест 2 → Тест 3 → Тест 4 → Тест 5.
- Если требуется выполнение нескольких тестов, дедлайн определяется по наиболее срочному тесту.

```python
from enum import Enum
from typing import List

# Определение перечислений для архитектуры, ролей и тестов
class Architecture(Enum):
    TYPE1 = "Тип 1"
    TYPE2 = "Тип 2"

class Role(Enum):
    DDNS = "DDNS"
    DNS = "DNS"
    GATEWAY = "Gateway"
    ROUTER = "Router"

class Test(Enum):
    TEST1 = "Тест 1"
    TEST2 = "Тест 2"
    TEST3 = "Тест 3"
    TEST4 = "Тест 4"
    TEST5 = "Тест 5"

# Функция для определения необходимых тестов в зависимости от архитектуры и ролей
def test_determine(architecture: Architecture, roles: List[Role]) -> List[Test]:
    tests = []

    if architecture == Architecture.TYPE1:
        tests.extend([Test.TEST1, Test.TEST2, Test.TEST3])  # Тип 1 требует Тест 1, 2, 3

    if architecture == Architecture.TYPE2:
        if Role.DDNS in roles:
            tests.extend([Test.TEST2, Test.TEST3])  # DDNS требует Тест 2, 3
        if Role.DNS in roles:
            tests.extend([Test.TEST4, Test.TEST5])  # DNS требует Тест 4, 5
        if Role.GATEWAY in roles:
            tests.extend([Test.TEST3, Test.TEST4])  # Шлюз требует Тест 3, 4
        if Role.ROUTER in roles:
            tests.extend([Test.TEST1, Test.TEST3])  # Маршрутизатор требует Тест 1, 3

    # Убираем дубли и возвращаем уникальные тесты
    return list(set(tests))

# Функция для определения дедлайна в зависимости от выбранных тестов
def deadline_determine(tests: List[Test]) -> int:
    # Приоритет: Тест 1 -> Тест 2 -> Тест 3 -> Тест 4 -> Тест 5
    if Test.TEST1 in tests:
        return 3
    if Test.TEST2 in tests:
        return 7
    if Test.TEST3 in tests:
        return 10
    if Test.TEST4 in tests:
        return 12
    if Test.TEST5 in tests:
        return 14
    return 0  # Если не выбраны тесты

# Основная функция для симуляции взаимодействия с пользователем
def main():
    # Bыбор архитектуры от пользователя
    print("Выберите архитектуру:")
    print("1. Тип 1")
    print("2. Тип 2")
    architecture_choice = int(input("Введите выбор: "))
    architecture = Architecture.TYPE1 if architecture_choice == 1 else Architecture.TYPE2

    # Bыбор ролей от пользователя
    print("Выберите роли (через запятую, например, DDNS,GATEWAY):")
    print("- DDNS")
    print("- DNS")
    print("- GATEWAY")
    print("- ROUTER")
    roles_input = input("Введите выбор: ").strip().lower().split(",")

    roles = []
    for role_str in roles_input:
        role = role_str.strip().upper()
        if role == 'DDNS':
            roles.append(Role.DDNS)
        elif role == 'DNS':
            roles.append(Role.DNS)
        elif role == 'GATEWAY':
            roles.append(Role.GATEWAY)
        elif role == 'ROUTER':
            roles.append(Role.ROUTER)

    # Tесты и дедлайн в зависимости от конфигурации
    chosen_test = test_determine(architecture, roles)
    deadline = deadline_determine(chosen_test)
#
    print(f"\nКонфигурация: Архитектура={architecture.value}, Роли={[role.value for role in roles]}")
    print(f"Выбранные тесты: {[test.value for test in chosen_test]}")
    print(f"Дедлайн: {deadline} дней")

if __name__ == "__main__":
    main()
```
